```python
"""
MesoNet + LSTM Architecture for Deepfake Detection using PyTorch
Optimized for Google Colab A100 GPU
"""

import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import datetime
import time
import random
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from collections import defaultdict

# PyTorch imports
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from torch.cuda.amp import autocast, GradScaler

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# For reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed()

# Global configs
BATCH_SIZE = 16
NUM_WORKERS = 4
FRAME_COUNT = 20  # Number of frames to extract from each video
IMAGE_SIZE = 128  # MesoNet typically uses 256x256, but smaller for faster training
LEARNING_RATE = 3e-4
NUM_EPOCHS = 30
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

##############################################
# Model Architecture: MesoNet + LSTM
##############################################

class MesoNet(nn.Module):
    """
    Implementation of MesoNet model for deepfake detection
    """
    def __init__(self):
        super(MesoNet, self).__init__()

        # First conv block
        self.conv1 = nn.Conv2d(3, 8, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(8)
        self.pool1 = nn.MaxPool2d(2, 2)

        # Second conv block
        self.conv2 = nn.Conv2d(8, 16, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(16)
        self.pool2 = nn.MaxPool2d(2, 2)

        # Third conv block
        self.conv3 = nn.Conv2d(16, 32, 3, padding=1, bias=False)
        self.bn3 = nn.BatchNorm2d(32)
        self.pool3 = nn.MaxPool2d(2, 2)

        # Fourth conv block
        self.conv4 = nn.Conv2d(32, 64, 3, padding=1, bias=False)
        self.bn4 = nn.BatchNorm2d(64)
        self.pool4 = nn.MaxPool2d(2, 2)

        # Calculate feature dim based on input size
        feature_dim = 64 * (IMAGE_SIZE // 16) * (IMAGE_SIZE // 16)

        # Feature size after convolution
        self.feature_size = feature_dim

    def forward(self, x):
        # First block
        x = self.pool1(F.relu(self.bn1(self.conv1(x))))

        # Second block
        x = self.pool2(F.relu(self.bn2(self.conv2(x))))

        # Third block
        x = self.pool3(F.relu(self.bn3(self.conv3(x))))

        # Fourth block
        x = self.pool4(F.relu(self.bn4(self.conv4(x))))

        return x


class MesoNetLSTM(nn.Module):
    """
    Combined MesoNet + LSTM model for sequential frame analysis
    """
    def __init__(self, lstm_hidden_size=128, lstm_layers=2, dropout=0.5):
        super(MesoNetLSTM, self).__init__()

        # MesoNet base model for feature extraction
        self.mesonet = MesoNet()

        # Feature dim calculated from MesoNet's output
        feature_dim = self.mesonet.feature_size

        # Flatten features for LSTM input
        self.flatten = nn.Flatten()

        # LSTM layer
        self.lstm = nn.LSTM(
            input_size=feature_dim,
            hidden_size=lstm_hidden_size,
            num_layers=lstm_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if lstm_layers > 1 else 0
        )

        # Final classification layers
        self.dropout = nn.Dropout(dropout)
        self.fc1 = nn.Linear(lstm_hidden_size * 2, 64)  # *2 for bidirectional
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        # x shape: [batch, frames, channels, height, width]
        batch_size, seq_len = x.size(0), x.size(1)

        # Process each frame through MesoNet to extract features
        frame_features = []
        for t in range(seq_len):
            frame = x[:, t, :, :, :]  # Get frame at time t
            features = self.mesonet(frame)  # Extract features
            features = self.flatten(features)  # Flatten for LSTM
            frame_features.append(features)

        # Stack features for LSTM
        lstm_input = torch.stack(frame_features, dim=1)  # [batch, seq_len, features]

        # Process through LSTM
        lstm_out, _ = self.lstm(lstm_input)

        # Take only the output from the last time step
        lstm_out = lstm_out[:, -1, :]

        # Final classification
        x = self.dropout(lstm_out)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)

        return x  # Raw logits for BCEWithLogitsLoss


##############################################
# Data Processing
##############################################

def extract_frames(video_path, num_frames=FRAME_COUNT):
    """Extract frames from a video file"""
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"Error: Could not open video {video_path}")
        return None

    frames = []
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if frame_count <= 0:
        print(f"Error: No frames in video {video_path}")
        return None

    # Calculate frame indices to extract (evenly distributed)
    indices = np.linspace(0, frame_count - 1, num_frames, dtype=int)

    for idx in indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()

        if ret:
            # Convert to RGB (from BGR)
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            # Resize frame
            frame = cv2.resize(frame, (IMAGE_SIZE, IMAGE_SIZE))

            frames.append(frame)

    cap.release()

    # If we couldn't extract enough frames, repeat the last frame
    while len(frames) < num_frames:
        if len(frames) > 0:
            frames.append(frames[-1])
        else:
            # Create an empty black frame if no frames were extracted
            frames.append(np.zeros((IMAGE_SIZE, IMAGE_SIZE, 3), dtype=np.uint8))

    return np.array(frames)


class DeepfakeDataset(Dataset):
    """PyTorch Dataset for deepfake videos"""
    def __init__(self, df, base_path, transform=None, preload=False):
        self.df = df
        self.base_path = base_path
        self.transform = transform
        self.preload = preload
        self.frames_cache = {}

        # Preload frames if enabled
        if preload:
            for idx in tqdm(range(len(df)), desc="Preloading frames"):
                video_path = os.path.join(base_path, df.iloc[idx]['file_path'])
                self.frames_cache[idx] = extract_frames(video_path)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # Get frames
        if idx in self.frames_cache:
            frames = self.frames_cache[idx]
        else:
            video_path = os.path.join(self.base_path, self.df.iloc[idx]['file_path'])
            frames = extract_frames(video_path)

            if frames is None:
                # Return empty frames if extraction failed
                frames = np.zeros((FRAME_COUNT, IMAGE_SIZE, IMAGE_SIZE, 3), dtype=np.uint8)

        # Apply transforms
        if self.transform:
            transformed_frames = []
            for frame in frames:
                transformed_frames.append(self.transform(frame))
            frames = torch.stack(transformed_frames)

        # Get label
        label = 1 if self.df.iloc[idx]['label'] == 'FAKE' else 0

        return frames, torch.tensor([label], dtype=torch.float32)


##############################################
# Training and Evaluation Functions
##############################################

def train_epoch(model, dataloader, criterion, optimizer, device, scaler):
    """Train for one epoch"""
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    # Progress bar
    pbar = tqdm(dataloader, desc="Training")

    for frames, labels in pbar:
        # Move data to device
        frames = frames.to(device)
        labels = labels.to(device)

        # Zero gradients
        optimizer.zero_grad()

        # Forward pass with mixed precision
        with autocast():
            outputs = model(frames)
            loss = criterion(outputs, labels)

        # Backward pass with gradient scaling
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        # Update statistics
        running_loss += loss.item() * frames.size(0)
        predicted = (torch.sigmoid(outputs) > 0.5).float()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        # Update progress bar
        pbar.set_postfix({'loss': loss.item(), 'acc': correct/total})

    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc


def evaluate(model, dataloader, criterion, device):
    """Evaluate model on validation data"""
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for frames, labels in tqdm(dataloader, desc="Evaluating"):
            # Move data to device
            frames = frames.to(device)
            labels = labels.to(device)

            # Forward pass
            outputs = model(frames)
            loss = criterion(outputs, labels)

            # Update statistics
            running_loss += loss.item() * frames.size(0)
            predicted = (outputs > 0.5).float()
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Store predictions and labels for metrics
            all_preds.extend(outputs.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculate metrics
    epoch_loss = running_loss / total
    epoch_acc = correct / total

    return epoch_loss, epoch_acc, np.array(all_preds), np.array(all_labels)


def train_model(model, train_loader, val_loader, criterion, optimizer,
                scheduler, device, num_epochs, model_save_path):
    """Complete training process with tracking metrics"""
    best_val_acc = 0.0
    history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': []
    }

    # Create a gradient scaler for mixed precision training
    scaler = GradScaler()

    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")
        print('-' * 30)

        # Train
        train_loss, train_acc = train_epoch(model, train_loader, criterion,
                                           optimizer, device, scaler)

        # Evaluate
        val_loss, val_acc, val_preds, val_labels = evaluate(model, val_loader,
                                                          criterion, device)

        # Update learning rate
        scheduler.step(val_loss)

        # Print metrics
        print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
        print(f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")

        # Update history
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)

        # Save model if better than previous best
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            print(f"New best validation accuracy: {val_acc:.4f}")
            torch.save(model.state_dict(), model_save_path)
            print(f"Model saved to {model_save_path}")

    return history


##############################################
# Visualization Functions
##############################################

def plot_metrics(history, title):
    """Plot training and validation metrics"""
    plt.figure(figsize=(12, 4))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history['train_acc'], label='Training Accuracy')
    plt.plot(history['val_acc'], label='Validation Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history['train_loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(f'{title}_metrics.png')
    plt.show()


def plot_confusion_matrix(y_true, y_pred, title):
    """Plot confusion matrix"""
    # Convert probabilities to binary predictions
    y_pred_binary = (np.array(y_pred) > 0.5).astype(int)
    y_true = np.array(y_true).astype(int)

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred_binary)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{title} - Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks([0.5, 1.5], ['Real', 'Fake'])
    plt.yticks([0.5, 1.5], ['Real', 'Fake'])
    plt.tight_layout()
    plt.savefig(f'{title}_confusion_matrix.png')
    plt.show()

    # Print classification report
    print(f"\nClassification Report for {title}:")
    target_names = ['Real', 'Fake']
    print(classification_report(y_true, y_pred_binary, target_names=target_names))


def plot_roc_curve(y_true, y_score, title):
    """Plot ROC curve"""
    y_true = np.array(y_true).astype(int)
    y_score = np.array(y_score)

    # Compute ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.3f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{title} - ROC Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f'{title}_roc_curve.png')
    plt.show()


##############################################
# Main Execution
##############################################

def main():
    # Set paths
    base_path = "/content/drive/MyDrive/Dataset-3"
    csv_path = os.path.join(base_path, "global_metadata_cleaned.csv")
    models_path = os.path.join(base_path, "models")

    # Create models directory if it doesn't exist
    os.makedirs(models_path, exist_ok=True)

    # Load metadata
    print("Loading metadata...")
    df = pd.read_csv(csv_path)

    # Data transforms
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    # Training with each dataset separately and then combined
    for dataset_config in [
        {'name': 'DFDC', 'datasets': ['DFDC']},
        {'name': 'FF', 'datasets': ['FF']},
        {'name': 'Combined', 'datasets': ['DFDC', 'FF']}
    ]:
        print(f"\n{'='*50}")
        print(f"Training on {dataset_config['name']} dataset")
        print(f"{'='*50}")

        # Filter data
        dataset_df = df[df['dataset'].isin(dataset_config['datasets'])]
        print(f"Total samples: {len(dataset_df)}")

        # Split train/test
        train_df = dataset_df[dataset_df['subset'] == 'train']
        test_df = dataset_df[dataset_df['subset'] == 'test']
        print(f"Train samples: {len(train_df)}, Test samples: {len(test_df)}")

        # Create datasets
        print("Creating datasets...")
        train_dataset = DeepfakeDataset(train_df, base_path, transform=transform)
        test_dataset = DeepfakeDataset(test_df, base_path, transform=transform)

        # Create dataloaders
        train_loader = DataLoader(
            train_dataset,
            batch_size=BATCH_SIZE,
            shuffle=True,
            num_workers=NUM_WORKERS,
            pin_memory=True
        )

        test_loader = DataLoader(
            test_dataset,
            batch_size=BATCH_SIZE,
            shuffle=False,
            num_workers=NUM_WORKERS,
            pin_memory=True
        )

        # Initialize model
        print("Initializing model...")
        model = MesoNetLSTM().to(DEVICE)
        print(f"Model on device: {DEVICE}")

        # Loss function and optimizer
        criterion = nn.BCEWithLogitsLoss()
        optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.5, patience=3, verbose=True
        )

        # Model save path
        timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
        model_save_path = os.path.join(models_path, f"mesonet_lstm_{dataset_config['name']}_{timestamp}.pth")

        # Train model
        print("Starting training...")
        history = train_model(
            model, train_loader, test_loader, criterion, optimizer,
            scheduler, DEVICE, NUM_EPOCHS, model_save_path
        )

        # Load best model for evaluation
        print("Loading best model for evaluation...")
        model.load_state_dict(torch.load(model_save_path))

        # Evaluate on test set
        _, test_acc, test_preds, test_labels = evaluate(model, test_loader, criterion, DEVICE)
        print(f"\nFinal Test Accuracy: {test_acc:.4f}")

        # Plot metrics
        plot_metrics(history, f"{dataset_config['name']}_model")

        # Plot confusion matrix
        plot_confusion_matrix(test_labels, test_preds, f"{dataset_config['name']}_model")

        # Plot ROC curve
        plot_roc_curve(test_labels, test_preds, f"{dataset_config['name']}_model")

        print(f"\nTraining and evaluation completed for {dataset_config['name']} dataset")
        print(f"Best model saved to {model_save_path}")


if __name__ == "__main__":
    main()
```

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import cv2
from torchvision import transforms

# Make sure these match your training config
FRAME_COUNT = 20
IMAGE_SIZE = 128
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ------------------------------
# Define MesoNet and MesoNetLSTM
# (Use your previously defined classes)
# ------------------------------
class MesoNet(nn.Module):
    def __init__(self):
        super(MesoNet, self).__init__()
        self.conv1 = nn.Conv2d(3, 8, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(8)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(8, 16, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(16)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(16, 32, 3, padding=1, bias=False)
        self.bn3 = nn.BatchNorm2d(32)
        self.pool3 = nn.MaxPool2d(2, 2)
        self.conv4 = nn.Conv2d(32, 64, 3, padding=1, bias=False)
        self.bn4 = nn.BatchNorm2d(64)
        self.pool4 = nn.MaxPool2d(2, 2)
        self.feature_size = 64 * (IMAGE_SIZE // 16) * (IMAGE_SIZE // 16)

    def forward(self, x):
        x = self.pool1(F.relu(self.bn1(self.conv1(x))))
        x = self.pool2(F.relu(self.bn2(self.conv2(x))))
        x = self.pool3(F.relu(self.bn3(self.conv3(x))))
        x = self.pool4(F.relu(self.bn4(self.conv4(x))))
        return x

class MesoNetLSTM(nn.Module):
    def __init__(self, lstm_hidden_size=128, lstm_layers=2, dropout=0.5):
        super(MesoNetLSTM, self).__init__()
        self.mesonet = MesoNet()
        feature_dim = self.mesonet.feature_size
        self.flatten = nn.Flatten()
        self.lstm = nn.LSTM(
            input_size=feature_dim,
            hidden_size=lstm_hidden_size,
            num_layers=lstm_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if lstm_layers > 1 else 0
        )
        self.dropout = nn.Dropout(dropout)
        self.fc1 = nn.Linear(lstm_hidden_size * 2, 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        batch_size, seq_len = x.size(0), x.size(1)
        frame_features = []
        for t in range(seq_len):
            features = self.mesonet(x[:, t, :, :, :])
            features = self.flatten(features)
            frame_features.append(features)
        lstm_input = torch.stack(frame_features, dim=1)
        lstm_out, _ = self.lstm(lstm_input)
        lstm_out = lstm_out[:, -1, :]
        x = self.dropout(lstm_out)
        x = F.relu(self.fc1(x))
        return self.fc2(x)
```

```python
def extract_frames(video_path, num_frames=FRAME_COUNT):
    cap = cv2.VideoCapture(video_path)
    frames = []
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    indices = np.linspace(0, total - 1, num_frames, dtype=int)

    for idx in indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (IMAGE_SIZE, IMAGE_SIZE))
        frames.append(frame)

    cap.release()
    while len(frames) < num_frames:
        frames.append(frames[-1])
    return np.array(frames)
```

```python
# Transformation
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

def predict(video_path):
    frames_np = extract_frames(video_path)
    frames_tensor = torch.stack([transform(frame) for frame in frames_np])
    frames_tensor = frames_tensor.unsqueeze(0).to(DEVICE)  # Shape: [1, seq_len, 3, H, W]

    with torch.no_grad():
        output = model(frames_tensor)
        prob = torch.sigmoid(output).item()
        label = "FAKE" if prob > 0.5 else "REAL"
    return prob, label

# Example
video = "/content/drive/MyDrive/Dataset-3/DFDC/test/FAKE/cglzoiprgl.mp4"
score, pred = predict(video)
print(f"Prediction: {pred} (Confidence: {score:.4f})")
```

```python
## ---- Combined Model ---- ##
## ---- Test Both Dataset --- ## (Test Folder only)

import os
import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from torchvision import transforms
from tqdm import tqdm

# Constants
IMAGE_SIZE = 128
FRAME_COUNT = 20
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_PATH = "/content/drive/MyDrive/Dataset-3/models/mesonet_lstm_Combined_20250521_193735.pth"
BASE_PATH = "/content/drive/MyDrive/Dataset-3"

# ---------- Model Definition ----------
class MesoNet(nn.Module):
    def __init__(self):
        super(MesoNet, self).__init__()
        self.conv1 = nn.Conv2d(3, 8, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(8)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(8, 16, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(16)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(16, 32, 3, padding=1, bias=False)
        self.bn3 = nn.BatchNorm2d(32)
        self.pool3 = nn.MaxPool2d(2, 2)
        self.conv4 = nn.Conv2d(32, 64, 3, padding=1, bias=False)
        self.bn4 = nn.BatchNorm2d(64)
        self.pool4 = nn.MaxPool2d(2, 2)
        self.feature_size = 64 * (IMAGE_SIZE // 16) * (IMAGE_SIZE // 16)

    def forward(self, x):
        x = self.pool1(F.relu(self.bn1(self.conv1(x))))
        x = self.pool2(F.relu(self.bn2(self.conv2(x))))
        x = self.pool3(F.relu(self.bn3(self.conv3(x))))
        x = self.pool4(F.relu(self.bn4(self.conv4(x))))
        return x

class MesoNetLSTM(nn.Module):
    def __init__(self, lstm_hidden_size=128, lstm_layers=2, dropout=0.5):
        super(MesoNetLSTM, self).__init__()
        self.mesonet = MesoNet()
        feature_dim = self.mesonet.feature_size
        self.flatten = nn.Flatten()
        self.lstm = nn.LSTM(
            input_size=feature_dim,
            hidden_size=lstm_hidden_size,
            num_layers=lstm_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if lstm_layers > 1 else 0
        )
        self.dropout = nn.Dropout(dropout)
        self.fc1 = nn.Linear(lstm_hidden_size * 2, 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        batch_size, seq_len = x.size(0), x.size(1)
        frame_features = []
        for t in range(seq_len):
            features = self.mesonet(x[:, t, :, :, :])
            features = self.flatten(features)
            frame_features.append(features)
        lstm_input = torch.stack(frame_features, dim=1)
        lstm_out, _ = self.lstm(lstm_input)
        lstm_out = lstm_out[:, -1, :]
        x = self.dropout(lstm_out)
        x = F.relu(self.fc1(x))
        return self.fc2(x)

# ---------- Load Model ----------
model = MesoNetLSTM().to(DEVICE)
model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
model.eval()

# ---------- Frame Extraction ----------
def extract_frames(video_path, num_frames=FRAME_COUNT):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    indices = np.linspace(0, total - 1, num_frames, dtype=int)
    frames = []

    for idx in indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (IMAGE_SIZE, IMAGE_SIZE))
        frames.append(frame)
    cap.release()

    while len(frames) < num_frames:
        frames.append(frames[-1])
    return np.array(frames)

# ---------- Transform ----------
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# ---------- Predict Single Video ----------
def predict(video_path):
    frames_np = extract_frames(video_path)
    frames_tensor = torch.stack([transform(frame) for frame in frames_np])
    frames_tensor = frames_tensor.unsqueeze(0).to(DEVICE)  # [1, seq_len, 3, H, W]
    with torch.no_grad():
        output = model(frames_tensor)
        prob = torch.sigmoid(output).item()
        label = "FAKE" if prob > 0.5 else "REAL"
    return prob, label

# ---------- Batch Prediction ----------
def collect_predictions(dataset_name, base_path):
    results = []
    test_path = os.path.join(base_path, dataset_name, "test")

    for true_label in ["REAL", "FAKE"]:
        folder = os.path.join(test_path, true_label)
        for fname in tqdm(os.listdir(folder), desc=f"{dataset_name} - {true_label}"):
            if not fname.endswith(".mp4"):
                continue
            try:
                video_path = os.path.join(folder, fname)
                prob, pred_label = predict(video_path)
                results.append({
                    "filename": fname,
                    "true_label": true_label,
                    "pred_label": pred_label,
                    "confidence": prob,
                    "dataset": dataset_name
                })
            except Exception as e:
                print(f"Error with {fname}: {e}")
    return pd.DataFrame(results)

# ---------- Run Inference ----------
df_dfdc = collect_predictions("DFDC", BASE_PATH)
df_ff = collect_predictions("FF", BASE_PATH)
df_all = pd.concat([df_dfdc, df_ff], ignore_index=True)

# ---------- Save Results ----------
df_all.to_csv(os.path.join(BASE_PATH, "inference_results.csv"), index=False)
print("Results saved to inference_results.csv")

# ---------- Visualization ----------
# Plot Predicted Label Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df_all, x="pred_label", hue="dataset")
plt.title("Predicted Label Distribution by Dataset")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot Confidence Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df_all, x="confidence", hue="true_label", bins=20, kde=True, element="step")
plt.title("Prediction Confidence by True Label")
plt.xlabel("Confidence Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()

# Accuracy Bar Chart
df_all["correct"] = df_all["true_label"] == df_all["pred_label"]
acc_by_dataset = df_all.groupby("dataset")["correct"].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=acc_by_dataset, x="dataset", y="correct")
plt.title("Accuracy by Dataset")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
## ---- COmbined Model ---- ##
## ---- Test Both Dataset --- ## (For whole dataset)

import os
import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from torchvision import transforms
from tqdm import tqdm

# Constants
IMAGE_SIZE = 128
FRAME_COUNT = 20
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_PATH = "/content/drive/MyDrive/Dataset-3/models/mesonet_lstm_Combined_20250521_193735.pth"
BASE_PATH = "/content/drive/MyDrive/Dataset-3"

# ---------- Model Definition ----------
class MesoNet(nn.Module):
    def __init__(self):
        super(MesoNet, self).__init__()
        self.conv1 = nn.Conv2d(3, 8, 3, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(8)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(8, 16, 3, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(16)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(16, 32, 3, padding=1, bias=False)
        self.bn3 = nn.BatchNorm2d(32)
        self.pool3 = nn.MaxPool2d(2, 2)
        self.conv4 = nn.Conv2d(32, 64, 3, padding=1, bias=False)
        self.bn4 = nn.BatchNorm2d(64)
        self.pool4 = nn.MaxPool2d(2, 2)
        self.feature_size = 64 * (IMAGE_SIZE // 16) * (IMAGE_SIZE // 16)

    def forward(self, x):
        x = self.pool1(F.relu(self.bn1(self.conv1(x))))
        x = self.pool2(F.relu(self.bn2(self.conv2(x))))
        x = self.pool3(F.relu(self.bn3(self.conv3(x))))
        x = self.pool4(F.relu(self.bn4(self.conv4(x))))
        return x

class MesoNetLSTM(nn.Module):
    def __init__(self, lstm_hidden_size=128, lstm_layers=2, dropout=0.5):
        super(MesoNetLSTM, self).__init__()
        self.mesonet = MesoNet()
        feature_dim = self.mesonet.feature_size
        self.flatten = nn.Flatten()
        self.lstm = nn.LSTM(
            input_size=feature_dim,
            hidden_size=lstm_hidden_size,
            num_layers=lstm_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if lstm_layers > 1 else 0
        )
        self.dropout = nn.Dropout(dropout)
        self.fc1 = nn.Linear(lstm_hidden_size * 2, 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        batch_size, seq_len = x.size(0), x.size(1)
        frame_features = []
        for t in range(seq_len):
            features = self.mesonet(x[:, t, :, :, :])
            features = self.flatten(features)
            frame_features.append(features)
        lstm_input = torch.stack(frame_features, dim=1)
        lstm_out, _ = self.lstm(lstm_input)
        lstm_out = lstm_out[:, -1, :]
        x = self.dropout(lstm_out)
        x = F.relu(self.fc1(x))
        return self.fc2(x)

# ---------- Load Model ----------
model = MesoNetLSTM().to(DEVICE)
model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
model.eval()

# ---------- Frame Extraction ----------
def extract_frames(video_path, num_frames=FRAME_COUNT):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    indices = np.linspace(0, total - 1, num_frames, dtype=int)
    frames = []

    for idx in indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (IMAGE_SIZE, IMAGE_SIZE))
        frames.append(frame)
    cap.release()

    while len(frames) < num_frames:
        frames.append(frames[-1])
    return np.array(frames)

# ---------- Transform ----------
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# ---------- Predict Single Video ----------
def predict(video_path):
    frames_np = extract_frames(video_path)
    frames_tensor = torch.stack([transform(frame) for frame in frames_np])
    frames_tensor = frames_tensor.unsqueeze(0).to(DEVICE)  # [1, seq_len, 3, H, W]
    with torch.no_grad():
        output = model(frames_tensor)
        prob = torch.sigmoid(output).item()
        label = "FAKE" if prob > 0.5 else "REAL"
    return prob, label

# ---------- Loop Over Full Dataset ----------
def collect_predictions_full(base_path, dataset_name):
    results = []
    for subset in ["train", "test"]:
        for true_label in ["REAL", "FAKE"]:
            folder = os.path.join(base_path, dataset_name, subset, true_label)
            if not os.path.exists(folder):
                continue
            for fname in tqdm(os.listdir(folder), desc=f"{dataset_name}/{subset}/{true_label}"):
                if not fname.endswith(".mp4"):
                    continue
                try:
                    video_path = os.path.join(folder, fname)
                    prob, pred_label = predict(video_path)
                    results.append({
                        "filename": fname,
                        "subset": subset,
                        "true_label": true_label,
                        "pred_label": pred_label,
                        "confidence": prob,
                        "dataset": dataset_name
                    })
                except Exception as e:
                    print(f"Error with {fname}: {e}")
    return pd.DataFrame(results)

# ---------- Run Inference on Full Dataset ----------
df_dfdc_full = collect_predictions_full(BASE_PATH, "DFDC")
df_ff_full = collect_predictions_full(BASE_PATH, "FF")
df_all = pd.concat([df_dfdc_full, df_ff_full], ignore_index=True)

# ---------- Save to CSV ----------
output_csv = os.path.join(BASE_PATH, "inference_results_full.csv")
df_all.to_csv(output_csv, index=False)
print(f"Full predictions saved to: {output_csv}")

# ---------- Visualizations ----------
# Count Plot: Predictions by Dataset
plt.figure(figsize=(8, 5))
sns.countplot(data=df_all, x="pred_label", hue="dataset")
plt.title("Predicted Labels (FULL Dataset)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Histogram: Confidence by True Label
plt.figure(figsize=(10, 6))
sns.histplot(data=df_all, x="confidence", hue="true_label", bins=20, kde=True, element="step")
plt.title("Prediction Confidence by True Label (FULL Dataset)")
plt.xlabel("Confidence Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()

# Accuracy by Dataset
df_all["correct"] = df_all["true_label"] == df_all["pred_label"]
acc_by_dataset = df_all.groupby("dataset")["correct"].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=acc_by_dataset, x="dataset", y="correct")
plt.title("Accuracy by Dataset (FULL Dataset)")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(True)
plt.tight_layout()
plt.show()
```
