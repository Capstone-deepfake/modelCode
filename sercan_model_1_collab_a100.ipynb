```python
#GPT 도움 받아 GPU 섰어 참고로 comment의 부분을 나 아니고 gpt 섰어
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, LSTM, Dense, Dropout, Flatten
from tensorflow.keras.layers import TimeDistributed, Input, Concatenate, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, TensorBoard
from tensorflow.keras.utils import to_categorical
from google.colab import drive

# Check for GPU and configure TensorFlow
print("Checking for GPU availability...")
physical_devices = tf.config.list_physical_devices('GPU')
if len(physical_devices) > 0:
    print(f"Found {len(physical_devices)} GPU(s). Using GPU for training.")
    # Configure TensorFlow to use memory growth to avoid allocating all GPU memory at once
    for device in physical_devices:
        tf.config.experimental.set_memory_growth(device, True)
    # Set mixed precision policy for faster training on A100
    tf.keras.mixed_precision.set_global_policy('mixed_float16')
    print("Mixed precision training enabled (float16)")
else:
    print("No GPU found. Training will be slow.")

# Mount Google Drive
drive.mount('/content/drive')

# Configuration parameters - optimized for A100 GPU
NUM_FRAMES = 20  # Number of frames to extract from each video
FRAME_HEIGHT = 112  # Increased resolution for better feature extraction
FRAME_WIDTH = 112   # Increased resolution for better feature extraction
BATCH_SIZE = 32     # Increased batch size for A100
EPOCHS = 30         # More epochs since we have powerful GPU
LR = 0.0001
USE_DATA_AUGMENTATION = True  # Data augmentation to prevent overfitting

# Path to your datasets in Google Drive
BASE_DRIVE_PATH = '/content/drive/MyDrive/celeb_ff'
METADATA_PATH = '/content/drive/MyDrive/celeb_ff/global_meta_data.csv'

# The global_meta_data.csv contains entries for both datasets
# We don't need to specify individual dataset paths as they will be
# referenced through the folder_path column in the CSV

def load_metadata(csv_path):
    """
    Load and process the metadata CSV file
    """
    df = pd.read_csv(csv_path)
    print(f"Loaded {len(df)} entries from metadata")
    return df

def extract_frames(video_path, num_frames=NUM_FRAMES):
    """
    Extract frames from a video file
    """
    if not os.path.exists(video_path):
        print(f"Error: Video file not found: {video_path}")
        return None

    cap = cv2.VideoCapture(video_path)
    frames = []
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames == 0:
        print(f"Error: Could not read frames from {video_path}")
        return None

    # Calculate frame indices to extract (evenly spaced)
    frame_indices = np.linspace(0, total_frames - 1, num_frames, dtype=int)

    for idx in frame_indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if ret:
            # Resize and normalize frame
            frame = cv2.resize(frame, (FRAME_WIDTH, FRAME_HEIGHT))
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(frame / 255.0)  # Normalize to [0, 1]

    cap.release()

    if len(frames) < num_frames:
        print(f"Warning: Could only extract {len(frames)} frames from {video_path}")
        # If we have at least 1 frame, duplicate the last frame to reach num_frames
        if len(frames) > 0:
            last_frame = frames[-1]
            while len(frames) < num_frames:
                frames.append(last_frame)
            return np.array(frames)
        return None

    return np.array(frames)

def apply_augmentation(frames):
    """
    Apply random augmentation to video frames
    Used to increase training data diversity
    """
    if not USE_DATA_AUGMENTATION:
        return frames

    augmented_frames = []

    # Apply the same augmentation to all frames in the sequence
    # This keeps temporal consistency

    # Random horizontal flip
    if np.random.random() > 0.5:
        frames = frames[:, :, ::-1, :]

    # Random brightness/contrast adjustment
    if np.random.random() > 0.5:
        brightness = np.random.uniform(0.8, 1.2)
        frames = np.clip(frames * brightness, 0, 1)

    # Random rotation (slight)
    if np.random.random() > 0.7:
        angle = np.random.uniform(-10, 10)
        for frame in frames:
            center = (FRAME_WIDTH // 2, FRAME_HEIGHT // 2)
            rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
            rotated_frame = cv2.warpAffine(frame, rotation_matrix, (FRAME_WIDTH, FRAME_HEIGHT))
            augmented_frames.append(rotated_frame)
        return np.array(augmented_frames)

    return frames

def prepare_data_generator(metadata_df, base_path, batch_size=BATCH_SIZE, augment=False):
    """
    Generator function that yields batches of video data and labels
    Updated to work with both celeb and ff++ datasets
    """
    n = len(metadata_df)
    indices = np.arange(n)

    while True:
        np.random.shuffle(indices)

        for start_idx in range(0, n, batch_size):
            batch_indices = indices[start_idx:start_idx + batch_size]
            batch_df = metadata_df.iloc[batch_indices]

            batch_x = []
            batch_y = []

            for _, row in batch_df.iterrows():
                # Construct full video path from base_path and folder_path in CSV
                video_path = os.path.join(base_path, row['folder_path'])
                frames = extract_frames(video_path)

                if frames is not None:
                    # Apply augmentation if enabled and in training mode
                    if augment and USE_DATA_AUGMENTATION:
                        frames = apply_augmentation(frames)

                    batch_x.append(frames)
                    # Convert label to numerical: 'real' -> 0, 'fake' -> 1
                    label = 1 if row['label'] == 'fake' else 0
                    batch_y.append(label)

            if batch_x:
                # Convert to appropriate data type for mixed precision
                batch_x = np.array(batch_x, dtype=np.float32)
                batch_y = to_categorical(np.array(batch_y), num_classes=2)
                yield batch_x, batch_y

def build_lstm_cnn_model():
    """
    Build an LSTM+CNN model for video classification
    Enhanced version for A100 GPU with deeper network
    """
    # Input shape: (NUM_FRAMES, HEIGHT, WIDTH, 3)
    inputs = Input(shape=(NUM_FRAMES, FRAME_HEIGHT, FRAME_WIDTH, 3))

    # CNN feature extraction for each frame - deeper network for A100
    # First Conv block
    cnn = TimeDistributed(Conv2D(64, (3, 3), padding='same', activation='relu'))(inputs)
    cnn = TimeDistributed(BatchNormalization())(cnn)
    cnn = TimeDistributed(Conv2D(64, (3, 3), padding='same', activation='relu'))(cnn)
    cnn = TimeDistributed(BatchNormalization())(cnn)
    cnn = TimeDistributed(MaxPooling2D((2, 2)))(cnn)
    cnn = TimeDistributed(Dropout(0.25))(cnn)

    # Second Conv block
    cnn = TimeDistributed(Conv2D(128, (3, 3), padding='same', activation='relu'))(cnn)
    cnn = TimeDistributed(BatchNormalization())(cnn)
    cnn = TimeDistributed(Conv2D(128, (3, 3), padding='same', activation='relu'))(cnn)
    cnn = TimeDistributed(BatchNormalization())(cnn)
    cnn = TimeDistributed(MaxPooling2D((2, 2)))(cnn)
    cnn = TimeDistributed(Dropout(0.25))(cnn)

    # Third Conv block
    cnn = TimeDistributed(Conv2D(256, (3, 3), padding='same', activation='relu'))(cnn)
    cnn = TimeDistributed(BatchNormalization())(cnn)
    cnn = TimeDistributed(Conv2D(256, (3, 3), padding='same', activation='relu'))(cnn)
    cnn = TimeDistributed(BatchNormalization())(cnn)
    cnn = TimeDistributed(MaxPooling2D((2, 2)))(cnn)
    cnn = TimeDistributed(Dropout(0.25))(cnn)

    # Fourth Conv block
    cnn = TimeDistributed(Conv2D(512, (3, 3), padding='same', activation='relu'))(cnn)
    cnn = TimeDistributed(BatchNormalization())(cnn)
    cnn = TimeDistributed(MaxPooling2D((2, 2)))(cnn)
    cnn = TimeDistributed(Dropout(0.25))(cnn)

    cnn = TimeDistributed(Flatten())(cnn)

    # LSTM to capture temporal patterns - increased capacity
    lstm = LSTM(512, return_sequences=True, recurrent_dropout=0.1)(cnn)
    lstm = BatchNormalization()(lstm)
    lstm = LSTM(256, recurrent_dropout=0.1)(lstm)
    lstm = BatchNormalization()(lstm)

    # Dense layers for classification
    dense = Dense(256, activation='relu')(lstm)
    dense = BatchNormalization()(dense)
    dense = Dropout(0.5)(dense)
    dense = Dense(128, activation='relu')(dense)
    dense = BatchNormalization()(dense)
    dense = Dropout(0.3)(dense)
    outputs = Dense(2, activation='softmax')(dense)

    model = Model(inputs=inputs, outputs=outputs)

    # Compile the model
    # Use mixed precision optimizer for A100
    optimizer = Adam(learning_rate=LR)

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(), tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
    )

    return model

def plot_training_history(history):
    """
    Plot the training and validation accuracy/loss
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot accuracy
    ax1.plot(history.history['accuracy'], label='Training Accuracy')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()

    # Plot loss
    ax2.plot(history.history['loss'], label='Training Loss')
    ax2.plot(history.history['val_loss'], label='Validation Loss')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()

    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/celeb_ff/training_history.png')
    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    """
    Plot the confusion matrix
    """
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Real', 'Fake'],
                yticklabels=['Real', 'Fake'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/celeb_ff/confusion_matrix.png')
    plt.show()

def visualize_dataset_samples():
    """
    Visualize a few sample frames from real and fake videos from both datasets
    """
    # Define paths for both datasets
    celeb_real_path = os.path.join(BASE_DRIVE_PATH, 'celeb/train/real')
    celeb_fake_path = os.path.join(BASE_DRIVE_PATH, 'celeb/train/fake')
    ff_real_path = os.path.join(BASE_DRIVE_PATH, 'ff++/train/real')
    ff_fake_path = os.path.join(BASE_DRIVE_PATH, 'ff++/train/fake')

    paths_to_check = [
        ("Celeb-DF Real", celeb_real_path),
        ("Celeb-DF Fake", celeb_fake_path),
        ("FF++ Real", ff_real_path),
        ("FF++ Fake", ff_fake_path)
    ]

    plt.figure(figsize=(15, 12))
    subplot_idx = 1

    for dataset_name, path in paths_to_check:
        if os.path.exists(path):
            videos = os.listdir(path)
            if videos:
                video_path = os.path.join(path, videos[0])
                cap = cv2.VideoCapture(video_path)
                ret, frame = cap.read()
                if ret:
                    plt.subplot(2, 2, subplot_idx)
                    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
                    plt.title(f"{dataset_name}: {videos[0]}")
                    plt.axis('off')
                cap.release()

        subplot_idx += 1

    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/celeb_ff/dataset_samples.png')
    plt.show()

    # Print dataset statistics
    print("\nDataset Statistics:")
    for dataset_name, path in paths_to_check:
        if os.path.exists(path):
            video_count = len(os.listdir(path))
            print(f"{dataset_name}: {video_count} videos")

def main():
    # Verify drive is mounted
    if not os.path.exists('/content/drive'):
        print("Google Drive not mounted. Please mount it first.")
        return

    # Check if dataset exists
    if not os.path.exists(BASE_DRIVE_PATH):
        print(f"Dataset not found at {BASE_DRIVE_PATH}. Please check the path.")
        return

    # Create directories for outputs
    save_dir = '/content/drive/MyDrive/celeb_ff/models'
    logs_dir = '/content/drive/MyDrive/celeb_ff/logs'
    results_dir = '/content/drive/MyDrive/celeb_ff/results'

    for dir_path in [save_dir, logs_dir, results_dir]:
        os.makedirs(dir_path, exist_ok=True)

    # Visualize some samples from the dataset
    print("Visualizing dataset samples...")
    visualize_dataset_samples()

    # Load metadata
    print(f"Loading metadata from {METADATA_PATH}...")
    metadata_df = load_metadata(METADATA_PATH)

    # Check label distribution
    print("Label distribution:")
    if 'label' in metadata_df.columns:
        print(metadata_df['label'].value_counts())

    # Check dataset distribution
    if 'dataset' in metadata_df.columns:
        print("\nDataset distribution:")
        print(metadata_df['dataset'].value_counts())
        print("\nDataset and label cross-tabulation:")
        print(pd.crosstab(metadata_df['dataset'], metadata_df['label']))

    # Split metadata by 'split' column if it exists, otherwise create splits
    if 'split' in metadata_df.columns:
        train_df = metadata_df[metadata_df['split'] == 'train']
        val_df = metadata_df[metadata_df['split'] == 'val']
        test_df = metadata_df[metadata_df['split'] == 'test']
    else:
        # If no 'split' column, create our own splits
        train_df, temp_df = train_test_split(metadata_df, test_size=0.3, stratify=metadata_df['label'], random_state=42)
        val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['label'], random_state=42)

    print(f"Train set: {len(train_df)} videos")
    print(f"Validation set: {len(val_df)} videos")
    print(f"Test set: {len(test_df)} videos")

    # Optional: Balance the dataset if needed
    # Uncomment if you have class imbalance issues
    """
    # Balance training data
    real_train = train_df[train_df['label'] == 'real']
    fake_train = train_df[train_df['label'] == 'fake']

    # Get the minority class count
    min_count = min(len(real_train), len(fake_train))

    # Balance by downsampling majority class
    if len(real_train) > len(fake_train):
        real_train = real_train.sample(min_count, random_state=42)
        train_df = pd.concat([real_train, fake_train])
    elif len(fake_train) > len(real_train):
        fake_train = fake_train.sample(min_count, random_state=42)
        train_df = pd.concat([real_train, fake_train])

    print(f"Balanced train set: {len(train_df)} videos")
    """

    # Create data generators with augmentation for training
    train_generator = prepare_data_generator(train_df, BASE_DRIVE_PATH, BATCH_SIZE, augment=True)
    val_generator = prepare_data_generator(val_df, BASE_DRIVE_PATH, BATCH_SIZE, augment=False)

    # Build model optimized for A100
    model = build_lstm_cnn_model()
    model.summary()

    # Enhanced callbacks for A100 training
    model_checkpoint = ModelCheckpoint(
        os.path.join(save_dir, 'best_model.h5'),
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    )

    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=7,  # Increased patience for more complex model
        restore_best_weights=True,
        verbose=1
    )

    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=3,
        min_lr=1e-6,
        verbose=1
    )

    tensorboard = TensorBoard(
        log_dir=logs_dir,
        histogram_freq=1,
        write_graph=True,
        update_freq='epoch'
    )

    callbacks = [
        model_checkpoint,
        early_stopping,
        reduce_lr,
        tensorboard
    ]

    # Calculate steps per epoch
    steps_per_epoch = max(1, len(train_df) // BATCH_SIZE)
    validation_steps = max(1, len(val_df) // BATCH_SIZE)

    print(f"Steps per epoch: {steps_per_epoch}, Validation steps: {validation_steps}")

    # Train model
    print("Starting model training with A100 GPU optimization...")
    history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=EPOCHS,
    validation_data=val_generator,
    validation_steps=validation_steps,
    callbacks=callbacks
)


    # Plot and save training history
    plot_training_history(history)

    # Save the final model - both HDF5 and SavedModel formats
    model.save(os.path.join(save_dir, 'final_model.h5'))
    model.save(os.path.join(save_dir, 'saved_model.keras'))

    print(f"Final model saved to {os.path.join(save_dir, 'final_model.h5')}")

    # More comprehensive evaluation on test set
    print("\nPerforming comprehensive evaluation on test set...")
    test_generator = prepare_data_generator(test_df, BASE_DRIVE_PATH, batch_size=1, augment=False)

    # First, evaluate with model.evaluate
    test_steps = min(len(test_df), 200)  # Evaluate on more test samples but still limit for time
    test_results = model.evaluate(test_generator, steps=test_steps, verbose=1)

    # Print metrics from evaluate
    metrics = ['Loss', 'Accuracy', 'AUC', 'Precision', 'Recall']
    for metric_name, value in zip(metrics, test_results):
        print(f"{metric_name}: {value:.4f}")

    # Collect predictions for detailed analysis
    y_true = []
    y_pred = []
    y_scores = []  # Store prediction probabilities for ROC analysis
    video_paths = []  # Store paths for analysis of failures

    print("\nGenerating predictions for detailed analysis...")
    for i in range(min(len(test_df), 200)):
        if i % 50 == 0:
            print(f"Processing test sample {i+1}/{min(len(test_df), 200)}")

        x_batch, y_batch = next(test_generator)
        pred = model.predict(x_batch, verbose=0)

        # Store original video path for error analysis
        video_paths.append(test_df.iloc[i]['folder_path'])

        y_true.append(np.argmax(y_batch[0]))
        y_pred.append(np.argmax(pred[0]))
        y_scores.append(pred[0][1])  # Store probability of "fake" class

    # Plot confusion matrix
    plot_confusion_matrix(y_true, y_pred)

    # Print detailed classification report
    print("\nDetailed Classification Report:")
    print(classification_report(y_true, y_pred, target_names=['Real', 'Fake']))

    # Save predictions for further analysis
    results_df = pd.DataFrame({
        'video_path': video_paths,
        'true_label': y_true,
        'predicted_label': y_pred,
        'fake_probability': y_scores
    })
    results_df.to_csv(os.path.join(results_dir, 'test_predictions.csv'), index=False)

    # Identify and analyze misclassified videos
    misclassified = results_df[results_df['true_label'] != results_df['predicted_label']]
    print(f"\nNumber of misclassified videos: {len(misclassified)} out of {len(results_df)}")

    if len(misclassified) > 0:
        print("\nSample of misclassified videos:")
        for i, (_, row) in enumerate(misclassified.head(5).iterrows()):
            print(f"Video: {row['video_path']}")
            print(f"True label: {'Fake' if row['true_label'] == 1 else 'Real'}")
            print(f"Predicted: {'Fake' if row['predicted_label'] == 1 else 'Real'}")
            print(f"Fake probability: {row['fake_probability']:.4f}")
            print("---")

    print(f"\nEvaluation complete. Results saved to {results_dir}")
    print("\nDeepfake detection model training and evaluation completed successfully!")

if __name__ == "__main__":
    main()
```
